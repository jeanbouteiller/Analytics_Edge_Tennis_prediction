```python
def replace_spaces_dashes(x):
    x = str(x).replace(" ","")
    x = x.replace("_","")
    return x
```

```python
odds = pd.DataFrame()
for year in range (2001,2021):
    df_date = pd.read_csv('odds_data/raw_odds/'+str(year)+'.csv')
    odds = pd.concat([odds,df_date],sort=False,ignore_index=True)
    
odds['Date'] = pd.to_datetime(odds['Date'])
odds['year'] = odds['Date'].dt.year
odds['Date']=odds['Date'].apply(lambda x:x.strftime('%Y-%m-%d'))

odds['winner_unique_id']=odds['Winner'].apply(lambda x:replace_spaces_dashes(x).lower())+'//'+odds['year'].apply(lambda x:str(x))
odds['loser_unique_id']=odds['Loser'].apply(lambda x:replace_spaces_dashes(x).lower())+'//'+odds['year'].apply(lambda x:str(x))
odds['game_unique_id']=odds['winner_unique_id']+odds['loser_unique_id']
odds
```

```python
games=pd.read_csv('jean_work/processed_data/dataset_v1.csv')
games['winner_unique_id']=games['winner'].apply(lambda x:replace_spaces_dashes(x))+'//'+games['year'].apply(lambda x:str(x))
games['loser_unique_id']=games['loser'].apply(lambda x:replace_spaces_dashes(x))+'//'+games['year'].apply(lambda x:str(x))
games['game_unique_id']=games['winner_unique_id']+games['loser_unique_id']
games = games[games['year']>2000]
games
```
