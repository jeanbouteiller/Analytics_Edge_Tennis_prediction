```python
def replace_spaces_dashes(x):
    x = str(x).replace(" ","")
    x = x.replace("_","")
    return x
```

```python
def change_round(Round):
    if Round=='1st Round':
        return 1
    if Round=='2nd Round':
        return 2
    if Round=='3rd Round':
        return 3
    if Round=='4th Round':
        return 4
    if Round=='Quarterfinals':
        return 5
    if Round=='Semifinals':
        return 6
    if Round=='The Final':
        return 7
```

```python
odds = pd.DataFrame()
for year in range (2001,2021):
    df_date = pd.read_csv('odds_data/raw_odds/'+str(year)+'.csv')
    odds = pd.concat([odds,df_date],sort=False,ignore_index=True)

odds['Date'] = pd.to_datetime(odds['Date'])
odds['year'] = odds['Date'].dt.year
odds['Date']=odds['Date'].apply(lambda x:x.strftime('%Y-%m-%d'))
odds['tournoi_unique_id'] = odds['year'].apply(lambda x:str(x))+'-'+odds['Location'].apply(lambda x:x.lower())+'-'+odds['Surface'].apply(lambda x:x.lower())
odds['Round']=odds['Round'].apply(lambda x:change_round(x))    
odds['tournoi_unique_id_round'] = odds['tournoi_unique_id'] + '-' + odds['Round'].apply(lambda x:str(x))
                                                                                           
odds['winner_unique_id']=odds['Winner'].apply(lambda x:replace_spaces_dashes(x).lower())+'//'+odds['year'].apply(lambda x:str(x))
odds['loser_unique_id']=odds['Loser'].apply(lambda x:replace_spaces_dashes(x).lower())+'//'+odds['year'].apply(lambda x:str(x))
odds['game_unique_id']=odds['winner_unique_id']+odds['loser_unique_id']
odds
```

```python
games=pd.read_csv('jean_work/processed_data/dataset_v1.csv')
games['tournoi_unique_id_round'] = games['tournoi_unique_id'] + '-' + games['round'].apply(lambda x:str(x))
games['winner_unique_id']=games['winner'].apply(lambda x:replace_spaces_dashes(x))+'//'+games['year'].apply(lambda x:str(x))
games['loser_unique_id']=games['loser'].apply(lambda x:replace_spaces_dashes(x))+'//'+games['year'].apply(lambda x:str(x))
games['game_unique_id']=games['winner_unique_id']+games['loser_unique_id']
games = games[games['year']>2000]
games
```

```python
merged = pd.merge(games,odds,on='game_unique_id')
```

```python
maxWmaxL = merged[["MaxW","MaxL"]].copy()
maxWmaxL = maxWmaxL.dropna()
maxWmaxL["oddsCorrect"]=maxWmaxL.apply(lambda row: row.MaxW<row.MaxL, axis = 1)
```

```python
maxWmaxL["oddsCorrect"].value_counts()[1]/len(maxWmaxL)
```

```python
odds1 = odds[["Date","MaxW","MaxL"]].copy()
odds1 = odds1.dropna()
odds1["Date"] = pd.to_datetime(odds1["Date"])
odds1["oddsCorrect"]=odds1.apply(lambda row: row.MaxW<row.MaxL, axis = 1)
odds1[odds1["Date"]<np.datetime64("2015-01-01")]["oddsCorrect"].value_counts()[1]/len(odds1[odds1["Date"]<np.datetime64("2015-01-01")])
```

```python
odds1['profit_bet_lowest'] = (odds1['oddsCorrect']*odds1['MaxW']*10 - 10)
odds1['cum_profit'] = odds1['profit_bet_lowest'].expanding().sum()
```

```python
odds1
```

```python
y = odds1[['Date','cum_profit']]
y["Date"] = pd.to_datetime(y["Date"])
y = y.set_index("Date")
```

```python
plt.title("Always betting 10$ per match on the lowest odd strategy")
plt.ylabel("Cumulated profit")
plt.plot(y)
```
