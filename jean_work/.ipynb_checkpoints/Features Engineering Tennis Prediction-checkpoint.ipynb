# First let's focus on nan values

1- Wrank and Lrank: we will look for the name of the payer in the ranking tab, and return his closest rank


```python
ranking_91_2020=pd.read_csv('ranking_average_all.csv')
ranking_91_2020['tourney_dates']
ranking_91_2020.head()
```

```python
#Define a function that returns the nearest date of a given date in a list of dates
def dist(date1,date2):
    date1=datetime.strptime(date1, '%Y-%m-%d')
    date2=datetime.strptime(date2, '%Y-%m-%d')
    return(abs((date1-date2).days))

def get_nearest_date(date,list_dates):
    min_dist=dist(date,list_dates[0])
    min_date=0
    for k in range(1,len(list_dates)):
        if dist(date,list_dates[k])<min_dist:
            min_dist=dist(date,list_dates[k])
            min_date=k
            if dist(date,list_dates[k])<=7:
                return list_dates[min_date]
    return list_dates[min_date]
            
def get_ranking(player_unique_id):
    player_name=player_unique_id.split('//')[0]
    date=player_unique_id.split('//')[1]
    ranking_player=ranking_91_2020[(ranking_91_2020['player_name']==player_name)]
    if len(ranking_player)>0:
        list_dates=ranking_player['tourney_dates'].values
        nearest_date=get_nearest_date(date,list_dates)
        return (ranking_91_2020[ranking_91_2020['unique_id']==player_name+nearest_date]['rank_number']).values[0]
    return np.nan
    
df_ranks_filled.loc[df_ranks_filled['wrank'].isna(),'wrank']=df_ranks_filled[df_ranks_filled['wrank'].isna()]['winner_unique_id'].apply(lambda x:get_ranking(x))
df_ranks_filled.loc[df_ranks_filled['lrank'].isna(),'lrank']=df_ranks_filled[df_ranks_filled['lrank'].isna()]['loser_unique_id'].apply(lambda x:get_ranking(x))
df_ranks_filled.to_csv('df_with_ranks_filled.csv',index=False)

```

2- Wpoints and Lpoints

```python
#We will then perform the same reasonning with ranking points
df_ranks_points_filled=pd.read_csv('df_with_ranks_filled.csv')

def get_points(player_unique_id):
    player_name=player_unique_id.split('//')[0]
    date=player_unique_id.split('//')[1]
    ranking_player=ranking_91_2020[(ranking_91_2020['player_name']==player_name)]
    if len(ranking_player)>0:
        list_dates=ranking_player['tourney_dates'].values
        nearest_date=get_nearest_date(date,list_dates)
        return (ranking_91_2020[ranking_91_2020['unique_id']==player_name+nearest_date]['ranking_points']).values[0]
    return np.nan

df_ranks_points_filled.loc[df_ranks_points_filled['wpts'].isna(),'wpts']=df_ranks_points_filled[df_ranks_points_filled['wpts'].isna()]['winner_unique_id'].apply(lambda x:get_points(x))
df_ranks_points_filled.loc[df_ranks_points_filled['lpts'].isna(),'lpts']=df_ranks_points_filled[df_ranks_points_filled['lpts'].isna()]['loser_unique_id'].apply(lambda x:get_points(x))
df_ranks_points_filled.to_csv('df_with_ranks_points_filled.csv',index=False)
```

# First use of algorithms

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

df_nan_filled=df_nan_filled.replace([np.inf, -np.inf], np.nan)
df_nan_filled=df_nan_filled.dropna()

df1_normalized=pd.get_dummies(df_nan_filled)
train_set, test_set = train_test_split(df1_normalized,test_size=0.2,random_state=42)

X_train=train_set.drop('winner_player',axis=1).reset_index(drop=True)
y_train=train_set['winner_player'].reset_index(drop=True)

X_test=test_set.drop('winner_player',axis=1).reset_index(drop=True)
y_test=test_set['winner_player'].reset_index(drop=True)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train=pd.DataFrame(ss.fit_transform(X_train))


X_test=pd.DataFrame(ss.fit_transform(X_test))


X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

n_max=25

cv_score_max=0
list_score=[]
for n in range (1,n_max):
    print(n)
    knn = KNeighborsClassifier(n_neighbors=n)
    cv_scores=cross_val_score(knn, X_train, y_train, cv=5)
    score_mean=float(np.mean(cv_scores))
    cv_score_max=max(cv_score_max,score_mean)
    list_score.append(score_mean)


plt.plot(list_score)
plt.xlabel('N_neighbours');
plt.ylabel('prediction_score')
plt.show();
print('Pour KNN, le score de prediction optimal est '+str(max(list_score))+' atteint pour n='+str(list_score.index(max(list_score))))
```
